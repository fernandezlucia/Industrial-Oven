```python
entrada = [] #entrada = [radioInterior, radioExterior, cantRadios, cantAngulos, iso, ninst]

#sacamos los datos de entrada
f = open('testExperimento1_1.in', 'r+')     
datos_iniciales = f.readline()
f.close()
datos_iniciales = datos_iniciales.split()
for i in datos_iniciales[:6]:
    entrada.append(int(i))

radioInterior = entrada[0]
radioExterior = entrada[1]
cantRadios = entrada[2]
cantAngulos = entrada[3]
iso = entrada[4]
ninst = entrada[5]

#print(entrada)

cantidad_puntos = cantAngulos*cantRadios
temperaturas_totales = []

#sacamos los datos de salida
g = open('temptestExperimento1_1.out','r') #vector de temperaturas respuesta de todas las instancias
datosSalida = g.read()
g.close()

datosSalida = datosSalida.splitlines(False)
for i in datosSalida[:len(datosSalida)]:
    temperaturas_totales.append(float(i))

#vector de vectores con los diferentes outputs de las instancias
instancias=[]
for i in range(0,len(temperaturas_totales),cantidad_puntos):
    instancia_i = temperaturas_totales[i: i + cantidad_puntos]
    instancias.append(instancia_i)


#vector de isotermnas de todas las instancias
g = open('isotestExperimento1_1.out','r') 
datosSalida = g.read()
g.close()

isotermas_totales = []
datosSalida = datosSalida.splitlines(False)
for i in datosSalida[:len(datosSalida)]:
    isotermas_totales.append(float(i))

#vector de vectores con las diferentes isotermnas de las instancias
isotermas = []     
for i in range(0,len(isotermas_totales),cantAngulos):
    isoterma_i = isotermas_totales[i: i + cantAngulos]
    isotermas.append(isoterma_i)

deltaR = (radioExterior - radioInterior)/(cantRadios-1)
radios = np.arange(radioInterior, radioExterior+deltaR, deltaR)

deltaTita = 2*np.pi/cantAngulos
angulos = np.arange(0, 2*np.pi, deltaTita)
angulos_deg = np.degrees(angulos)

inst = 0

temperatura = instancias[inst]
isoterma = isotermas[inst]

indice = 0
for i in range(0,cantAngulos):
  indice = int(isoterma[i])
  isoterma[i] = radios[indice]

radioPeligroso = int(cantRadios*(3/4))
radioPeligroso = radios[radioPeligroso]
radiosPeligrosos = []
for i in range(0, cantAngulos+1):
  radiosPeligrosos.append(radioPeligroso)

angulos = np.append(angulos, angulos[0])
isoterma = np.append(isoterma, isoterma[0])
ax = plt.subplot(111, polar=True)
ax.plot(angulos, isoterma, "b-")

ax.fill_between(np.linspace(0, 2*np.pi, 100), radioPeligroso-0.00005, radioPeligroso+0.00005, color='red', zorder=10) # <-- Added here
ax.set_rmin(0)
ax.set_rmax(radioExterior)
ax.set_thetagrids(angulos_deg)
ax.set_rticks([radioInterior, max(isoterma), min(isoterma), radioExterior, radioPeligroso])  
ax.set_title("Isoterma", pad=15)
plt.show
```
